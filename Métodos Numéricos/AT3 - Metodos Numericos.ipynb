### 1. Considere a funo f(x) = sen(2x) + cos(3x)

##### a) Obtenha o polinmio de Taylor de grau 2 [P2(x)], e de grau 3 [P3(x)], que aproximem a funco no ponto x = ;

#### Grfico de $f(x)$


1. Polinmio de Taylor de Segundo Grau:

    O polinmio de Taylor de segundo grau para $f(x)$ em $x = $  dado por:

    $P(x) = f() + f'()(x - ) + f''()(x - )/2$

    Substituindo $f(x), f'(x) e f''(x)$ com suas respectivas definies:

    $P(x) = sen(2) + 2cos(2)(x - ) - 4sen(2)(x - )/2$

    $P(x) = 0 + 2(-1)(x - ) - 4(0)(x - )/2$

    $P(x) = -2(x - )$





2. Polinmio de Taylor de Terceiro Grau:

    O polinmio de Taylor de terceiro grau para f(x) em x =   dado por:

    $P(x) = P(x) + f'''()(x - )/6$

    Substituindo $f'''() = 12cos(2) = 12$:

    $P(x) = -2(x - ) + 12(x - )/6$

    $P(x) = -2(x - ) + 2(x - )$



##### Cdigo em Python

```python
import numpy as np
import math
import matplotlib.pyplot as plt

a = np.pi

def fx(x):
    return np.sin(2 * x) + np.cos(3 * x) 

def fx1(x):
    return 2 * np.cos(2 * x) -3 * np.sin(3 * x) 

def fx2(x):
    return -4 * np.sin(2 * x) -9 * np.cos(3 * x) 

def fx3(x):
    return -8 * np.cos(2 * x) -27 * np.sin(3 * x) 

def pol_taylor_G2(x):
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2)

def pol_taylor_G3(x):
    #return -2 * (x - np.pi) + 2 * (x - np.pi) ** 3
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2) + (fx3(a) * (x - a) ** 3) / math.factorial(3)
```

### 2. Dada a funo $f(x) = ln(x + 1)$, e os pontos x0 =0,0; x1 =0,6; x2=0,9 e x3=1,1. Construa os polinmios de interpolaco de Lagrange nos seguintes casos:

Para construir os polinmios de interpolao de Lagrange, utilizamos a seguinte frmula geral para um polinmio $ P(x) $ de grau $ n $:

$P(x) = \sum_{i=0}^{n} y_i \cdot L_i(x)$

onde $ L_i(x) $ so os polinmios bsicos de Lagrange definidos por:

$ L_i(x) = \prod_{\substack{0 \le j \le n \\ j \ne i}} \frac{x - x_j}{x_i - x_j} $

Os pontos dados so:
- $ x_0 = 0,0 $
- $ x_1 = 0,6 $
- $ x_2 = 0,9 $
- $ x_3 = 1,1 $

A funo dada  $ f(x) = \ln(x + 1) $. Portanto, os valores de $ f(x) $ nos pontos dados so:
- $ f(x_0) = \ln(0,0 + 1) = \ln(1) = 0 $
- $ f(x_1) = \ln(0,6 + 1) = \ln(1,6) \approx 0,470 $
- $ f(x_2) = \ln(0,9 + 1) = \ln(1,9) \approx 0,641 $
- $ f(x_3) = \ln(1,1 + 1) = \ln(2,1) \approx 0,742 $

### a) Polinmio de grau 1, com os pontos $ x_0 $ e $ x_3 $

Para um polinmio de grau 1 utilizando $ x_0 $ e $ x_3 $:

$ P_1(x) = y_0 \cdot L_0(x) + y_3 \cdot L_3(x) $

onde:

$ L_0(x) = \frac{x - x_3}{x_0 - x_3} = \frac{x - 1,1}{0 - 1,1} = -\frac{x - 1,1}{1,1} $

$ L_3(x) = \frac{x - x_0}{x_3 - x_0} = \frac{x - 0}{1,1 - 0} = \frac{x}{1,1} $

Ento:

$ P_1(x) = 0 \cdot L_0(x) + 0,742 \cdot L_3(x)$

$ P_1(x) = 0,742 \cdot \frac{x}{1,1} = \frac{0,742}{1,1}x $

$ P_1(x) = 0,6745x $

### b) Polinmio de grau 2, com os pontos $ x_0 $, $ x_1 $ e $ x_3 $

Para um polinmio de grau 2 utilizando $ x_0 $, $ x_1 $ e $ x_3 $:

$ P_2(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_3 \cdot L_3(x) $

onde:

$ L_0(x) = \frac{(x - x_1)(x - x_3)}{(x_0 - x_1)(x_0 - x_3)} = \frac{(x - 0,6)(x - 1,1)}{(0 - 0,6)(0 - 1,1)} = \frac{(x - 0,6)(x - 1,1)}{0,66} $

$ L_1(x) = \frac{(x - x_0)(x - x_3)}{(x_1 - x_0)(x_1 - x_3)} = \frac{(x - 0)(x - 1,1)}{(0,6 - 0)(0,6 - 1,1)} = \frac{x(x - 1,1)}{-0,3} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)}{(x_3 - x_0)(x_3 - x_1)} = \frac{(x - 0)(x - 0,6)}{(1,1 - 0)(1,1 - 0,6)} = \frac{x(x - 0,6)}{0,55} $

Os valores de $ y_i $ so:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_3 = 0,742 $

Substituindo, obtemos:

$ P_2(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:

$ P_2(x) = 0 + 0,470 \cdot \frac{x(x - 1,1)}{-0,3} + 0,742 \cdot \frac{x(x - 0,6)}{0,55} $

### c) Polinmio de grau 3, com os pontos $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $

Para um polinmio de grau 3 utilizando $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $:

$ P_3(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_2 \cdot L_2(x) + y_3 \cdot L_3(x) $

onde:
$ L_0(x) = \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)} $

$ L_1(x) = \frac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)} $

$ L_2(x) = \frac{(x - x_0)(x - x_1)(x - x_3)}{(x_2 - x_0)(x_2 - x_1)(x_2 - x_3)} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)(x - x_2)}{(x_3 - x_0)(x_3 - x_1)(x_3 - x_2)} $

Os valores de $ y_i $ so:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_2 = 0,641 $
- $ y_3 = 0,742 $

Substituindo, obtemos:
$ P_3(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:
$ P_3(x) = 0 + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

### Polinmios Bsicos de Lagrange

Os polinmios bsicos $ L_i(x) $ para $ P_3(x) $ so:

- $ L_1(x) = \frac{(x - 0)(x - 0,9)(x - 1,1)}{(0,6 - 0)(0,6 - 0,9)(0,6 - 1,1)} $
- $ L_2(x) = \frac{(x - 0)(x - 0,6)(x - 1,1)}{(0,9 - 0)(0,9 - 0,6)(0,9 - 1,1)} $
- $ L_3(x) = \frac{(x - 0)(x - 0,6)(x - 0,9)}{(1,1 - 0)(1,1 - 0,6)(1,1 - 0,9)} $

Substituindo os valores especficos:
- $ L_1(x) = \frac{(x)(x - 0,9)(x - 1,1)}{0,6 \cdot (-0,3) \cdot (-0,5)} = \frac{(x)(x - 0,9)(x - 1,1)}{0,09} $
- $ L_2(x) = \frac{(x)(x - 0,6)(x - 1,1)}{0,9 \cdot 0,3 \cdot (-0,2)} = \frac{(x)(x - 0,6)(x - 1,1)}{-0,054} $
- $ L_3(x) = \frac{(x)(x - 0,6)(x - 0,9)}{1,1 \cdot 0,5 \cdot 0,2} = \frac{(x)(x - 0,6)(x - 0,9)}{0,11} $

Portanto:
$ P_3(x) = 0 + 0,470 \cdot \frac{(x)(x - 0,9)(x - 1,1)}{0,09} + 0,641 \cdot \frac{(x)(x - 0,6)(x - 1,1)}{-0,054} + 0,742 \cdot \frac{(x)(x - 0,6)(x - 0,9)}{0,11} $

#### Construa um grfico com a funo f(x), os pontos de interpolao e os polinmios P1, P2 e P3. Utilize os polinmios de interpolaco para avaliar a funo nos pontos x = 0,3 e x = 0,75. Calcule o desvio relativo com relao ao valor verdadeiro em cada caso. Comente os resultados obtidos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Funo original
def f(x):
    return np.log(x + 1)

# Pontos dados
x0, x1, x2, x3 = 0.0, 0.6, 0.9, 1.1
y0, y1, y2, y3 = f(x0), f(x1), f(x2), f(x3)

# Polinmio de grau 1
def P1(x):
    return 0.6745 * x

# Polinmio de grau 2
def P2(x):
    return 0.470 * x * (x - 1.1) / - 0.3 + 0.742 * x * (x - 0.6) / 0.55

# Polinmio de grau 3
def P3(x):
    return (0.470 * x * (x - 0.9) * (x - 1.1) / 0.09 +
            0.641 * x * (x - 0.6) * (x - 1.1) / - 0.054 +
            0.742 * x * (x - 0.6) * (x - 0.9) / 0.11)

# Valores de x para plotagem
x_values = np.linspace(-0.1, 1.2, 400)

# Valores de f(x)
y_values = f(x_values)

# Valores dos polinmios
P1_values = P1(x_values)
P2_values = P2(x_values)
P3_values = P3(x_values)

# Pontos de interpolao
x_points = [x0, x1, x2, x3]
y_points = [y0, y1, y2, y3]

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = ln(x + 1)', color='black')
plt.plot(x_values, P1_values, label='P1(x)', linestyle='--')
plt.plot(x_values, P2_values, label='P2(x)', linestyle='--')
plt.plot(x_values, P3_values, label='P3(x)', linestyle='--')
plt.scatter(x_points, y_points, label='Dados', color='red', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolao de Lagrange')
plt.grid(True)
plt.show()
```

```python
# Avaliao nos pontos x = 0.3 e x = 0.75
x_teste = [0.3, 0.75]
valores_f = f(np.array(x_teste))
P1_teste = P1(np.array(x_teste))
P2_teste = P2(np.array(x_teste))
P3_teste = P3(np.array(x_teste))

# Desvios relativos
desvio_relativo_P1 = np.abs((P1_teste - valores_f) / valores_f) * 100
desvio_relativo_P2 = np.abs((P2_teste - valores_f) / valores_f) * 100
desvio_relativo_P3 = np.abs((P3_teste - valores_f) / valores_f) * 100

# Resultados
print("X de teste:", x_teste)
print("Valores verdadeiros de f(x):", valores_f)
print("Polinminio de Grau 1:", P1_teste)
print("Polinminio de Grau 2:", P2_teste)
print("Polinminio de Grau 3:", P3_teste)
print("Desvio relativo P1 (%):", np.round(desvio_relativo_P1, 2))
print("Desvio relativo P2 (%):", np.round(desvio_relativo_P2, 2))
print("Desvio relativo P3 (%):", np.round(desvio_relativo_P3, 2))
```

Comentrios sobre os resultados:

- Polinmio de grau 1 (P1 (x)): O desvio relativo  significativamente maior em ambos os pontos (22.87% para x=0.3 e 9.60% para x=0.75). Isso mostra que o polinmio de grau 1 no  suficiente para uma boa aproximao da funo f(x), especialmente em pontos distantes dos ns de interpolao. 

- Polinmio de grau 2 (P2  (x)): O desvio relativo  muito menor em comparao ao polinmio de grau 1 (2.96% para x=0.3 e 0.61% para x=0.75). Isso indica que o polinmio de grau 2 proporciona uma aproximao significativamente melhor da funo f(x). 

- Polinmio de grau 3 (P3 (x)): O desvio relativo  extremamente pequeno (0.74% para x=0.3 e 0.04% para x=0.75), o que sugere que o polinmio de grau 3  capaz de capturar muito bem a forma da funo f(x), fornecendo uma aproximao quase exata nos pontos avaliados.
