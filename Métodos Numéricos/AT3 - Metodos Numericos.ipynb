### 1. Considere a função $f(x) = sen(2x) + cos(3x)$

#### a) Obtenha o polinômio de Taylor de grau 2 [P2(x)], e de grau 3 [P3(x)], que aproximem a funcão no ponto x = π;

#### Gráfico de $f(x)$


1. Polinômio de Taylor de Segundo Grau:

    O polinômio de Taylor de segundo grau para $f(x)$ em $x = π$ é dado por:

    $P₂(x) = f(π) + f'(π)(x - π) + f''(π)(x - π)²/2$

    Substituindo $f(x), f'(x) e f''(x)$ com suas respectivas definições:

    $P₂(x) = sen(2π) + 2cos(2π)(x - π) - 4sen(2π)(x - π)²/2$

    $P₂(x) = 0 + 2(-1)(x - π) - 4(0)(x - π)²/2$

    $P₂(x) = -2(x - π)$





2. Polinômio de Taylor de Terceiro Grau:

    O polinômio de Taylor de terceiro grau para f(x) em x = π é dado por:

    $P₃(x) = P₂(x) + f'''(π)(x - π)³/6$

    Substituindo $f'''(π) = 12cos(2π) = 12$:

    $P₃(x) = -2(x - π) + 12(x - π)³/6$

    $P₃(x) = -2(x - π) + 2(x - π)³$



##### Código em Python

```python
import numpy as np
import math
import matplotlib.pyplot as plt

a = np.pi

def fx(x):
    return np.sin(2 * x) + np.cos(3 * x) 

def fx1(x):
    return 2 * np.cos(2 * x) -3 * np.sin(3 * x) 

def fx2(x):
    return -4 * np.sin(2 * x) -9 * np.cos(3 * x) 

def fx3(x):
    return -8 * np.cos(2 * x) -27 * np.sin(3 * x) 

def pol_taylor_G2(x):
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2)

def pol_taylor_G3(x):
    #return -2 * (x - np.pi) + 2 * (x - np.pi) ** 3
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2) + (fx3(a) * (x - a) ** 3) / math.factorial(3)
```

### 2. Dada a função $f(x) = ln(x + 1)$, e os pontos $x_0 =0,0; $ $x_1 =0,6; $ $x_2=0,9$ e $x_3=1,1$. Construa os polinômios de interpolacão de Lagrange nos seguintes casos:

Para construir os polinômios de interpolação de Lagrange, utilizamos a seguinte fórmula geral para um polinômio $ P(x) $ de grau $ n $:

$P(x) = \sum_{i=0}^{n} f(x_i) \cdot L_i(x)$

onde $ L_i(x) $ são os polinômios básicos de Lagrange definidos por:

$ L_i(x) = \prod_{\substack{0 \le j \le n \\ j \ne i}} \frac{x - x_j}{x_i - x_j} $

Os pontos dados são:
- $ x_0 = 0,0 $
- $ x_1 = 0,6 $
- $ x_2 = 0,9 $
- $ x_3 = 1,1 $

A função dada é $ f(x) = \ln(x + 1) $. Portanto, os valores de $ f(x) $ nos pontos dados são:
- $ f(x_0) = \ln(0,0 + 1) = \ln(1) = 0 $
- $ f(x_1) = \ln(0,6 + 1) = \ln(1,6) \approx 0,470 $
- $ f(x_2) = \ln(0,9 + 1) = \ln(1,9) \approx 0,641 $
- $ f(x_3) = \ln(1,1 + 1) = \ln(2,1) \approx 0,742 $

#### a) Polinômio de grau 1, com os pontos $ x_0 $ e $ x_3 $

Para um polinômio de grau 1 utilizando $ x_0 $ e $ x_3 $:

$ P_1(x) = y_0 \cdot L_0(x) + y_3 \cdot L_3(x) $

onde:

$ L_0(x) = \frac{x - x_3}{x_0 - x_3} = \frac{x - 1,1}{0 - 1,1} = -\frac{x - 1,1}{1,1} $

$ L_3(x) = \frac{x - x_0}{x_3 - x_0} = \frac{x - 0}{1,1 - 0} = \frac{x}{1,1} $

Então:

$ P_1(x) = 0 \cdot L_0(x) + 0,742 \cdot L_3(x)$

$ P_1(x) = 0,742 \cdot \frac{x}{1,1} = \frac{0,742}{1,1}x $

$ P_1(x) = 0,6745x $

### b) Polinômio de grau 2, com os pontos $ x_0 $, $ x_1 $ e $ x_3 $

Para um polinômio de grau 2 utilizando $ x_0 $, $ x_1 $ e $ x_3 $:

$ P_2(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_3 \cdot L_3(x) $

onde:

$ L_0(x) = \frac{(x - x_1)(x - x_3)}{(x_0 - x_1)(x_0 - x_3)} = \frac{(x - 0,6)(x - 1,1)}{(0 - 0,6)(0 - 1,1)} = \frac{(x - 0,6)(x - 1,1)}{0,66} $

$ L_1(x) = \frac{(x - x_0)(x - x_3)}{(x_1 - x_0)(x_1 - x_3)} = \frac{(x - 0)(x - 1,1)}{(0,6 - 0)(0,6 - 1,1)} = \frac{x(x - 1,1)}{-0,3} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)}{(x_3 - x_0)(x_3 - x_1)} = \frac{(x - 0)(x - 0,6)}{(1,1 - 0)(1,1 - 0,6)} = \frac{x(x - 0,6)}{0,55} $

Os valores de $ y_i $ são:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_3 = 0,742 $

Substituindo, obtemos:

$ P_2(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:

$ P_2(x) = 0 + 0,470 \cdot \frac{x(x - 1,1)}{-0,3} + 0,742 \cdot \frac{x(x - 0,6)}{0,55} $

### c) Polinômio de grau 3, com os pontos $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $

Para um polinômio de grau 3 utilizando $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $:

$ P_3(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_2 \cdot L_2(x) + y_3 \cdot L_3(x) $

onde:
$ L_0(x) = \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)} $

$ L_1(x) = \frac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)} $

$ L_2(x) = \frac{(x - x_0)(x - x_1)(x - x_3)}{(x_2 - x_0)(x_2 - x_1)(x_2 - x_3)} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)(x - x_2)}{(x_3 - x_0)(x_3 - x_1)(x_3 - x_2)} $

Os valores de $ y_i $ são:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_2 = 0,641 $
- $ y_3 = 0,742 $

Substituindo, obtemos:
$ P_3(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:
$ P_3(x) = 0 + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

##### Polinômios Básicos de Lagrange

Os polinômios básicos $ L_i(x) $ para $ P_3(x) $ são:

Portanto:
$ P_3(x) = 0 + 0,470 \cdot \frac{(x)(x - 0,9)(x - 1,1)}{0,09} + 0,641 \cdot \frac{(x)(x - 0,6)(x - 1,1)}{-0,054} + 0,742 \cdot \frac{(x)(x - 0,6)(x - 0,9)}{0,11} $

### 3. A função de Runge é definida como f(x) = 1/(1 + x²).

#### a) Desejamos construir o polinômio interpolante de Lagrange no intervalo [−5, 5] com pontos equidistantes para n = 4, 6, 8 e 12.

O polinômio interpolante de Lagrange $P(x)$ é uma combinação linear dos valores da função $f$ nos pontos de interpolação, ponderada pelos polinômios base de Lagrange $l_i(x)$. A fórmula geral do polinômio de Lagrange para n pontos $(x_0, f(x_0)) , (x_1, f(x_1)),…,(x_{n-1},f(x_{n-1}))$ é dada por: 

$P(x) = \sum_{i=0}^{n-1} y_i \cdot L_i(x)$

onde $ L_i(x) $ são os polinômios básicos de Lagrange definidos por:

$ L_i(x) = \prod_{\substack{0 \le j \le n \\ j \ne i}} \frac{x - x_j}{x_i - x_j} $

Esta fórmula significa que cada $l_i(x)$ é um polinômio que é igual a 1 no ponto $x_i$ e igual a 0 nos demais pontos $x_j​$ (para $j \ne i$). Portanto, o polinômio de Lagrange $P(x)$ passa exatamente pelos pontos de interpolação fornecidos. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função de Runge
def funcao_runge(x):
    return 1 / (1 + x**2)

# Definindo os pontos equidistantes e valores da função
def pontos_equidistantes(a, b, n):
    return np.linspace(a, b, n)

# Calculando os polinômios de Lagrange
def lagrange_basicos(xi, x):
    n = len(xi)
    L = np.ones((n, len(x)))
    for i in range(n):
        for j in range(n):
            if i != j:
                L[i] *= (x - xi[j]) / (xi[i] - xi[j])
    return L

# Construindo o polinômio interpolante de Lagrange
def polinomio_interpolante_lagrange(xi, fi, x):
    L = lagrange_basicos(xi, x)
    P = np.dot(fi, L)
    return P
```

#### b) Em cada caso construa um gráfico contendo a função de Runge, os pontos de interpolação e o Polinômio de Lagrange.

```python
# Intervalo e número de pontos
a, b = -5, 5
pontos = [4, 6, 8, 12]
x_plot = np.linspace(a, b, 1000)

# Plotando a função original e os polinômios interpolantes para cada n
for n in pontos:
    xi = pontos_equidistantes(a, b, n)
    fi = funcao_runge(xi)
    P = polinomio_interpolante_lagrange(xi, fi, x_plot)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_plot, funcao_runge(x_plot), label='Função de Runge', color='black')
    plt.plot(x_plot, P, label=f'Polinômio de Lagrange (n={n})')
    plt.scatter(xi, fi, color='red')  # pontos de interpolação
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Polinômio de Lagrange Interpolante (n={n})')
    plt.grid(True)
    plt.show()
```

Os gráficos dos polinômios interpolantes de Lagrange para a função de Runge com $n=4,6, 8, 12$ pontos igualmente espaçados mostram alguns comportamentos típicos da interpolação polinomial, especialmente em relação à acurácia no centro do intervalo e nos extremos. Vamos analisar essas características mais detalhadamente:

##### Acurácia no Centro do Intervalo

No centro do intervalo $[−5,5]$, os polinômios interpolantes de Lagrange geralmente fornecem uma boa aproximação da função de Runge. Para $n=4, 6 ,8 e 12$ os polinômios interpolantes seguem bem a forma da função original, com a acurácia melhorando à medida que aumentamos o número de pontos. Este comportamento ocorre porque:

- Menores Erros: A interpolação polinomial tende a ser mais precisa próximo aos pontos de interpolação. No centro do intervalo, os pontos de interpolação estão mais próximos uns dos outros, reduzindo os erros de interpolação.
- Menor Oscilação: A oscilação (fenômeno de Runge) é menos pronunciada no centro do intervalo, onde os pontos são mais uniformemente distribuídos em relação ao ponto avaliado.
  
##### Acurácia nos Extremos do Intervalo

Nos extremos do intervalo $[−5,5]$, a acurácia da interpolação polinomial de Lagrange diminui significativamente, especialmente com o aumento do número de pontos de interpolação. Observamos oscilações pronunciadas nos gráficos para $n=8 e 12$. Este comportamento pode ser explicado por:

- Fenômeno de Runge: Este fenômeno descreve a tendência dos polinômios interpolantes de alta ordem de oscilar fortemente nas extremidades do intervalo de interpolação. À medida que aumentamos o grau do polinômio, a oscilação nas extremidades se intensifica.
- Distribuição dos Pontos: Em um conjunto de pontos igualmente espaçados, a densidade de pontos de interpolação é menor nas extremidades do intervalo. Isso pode levar a grandes valores nos denominadores dos polinômios de base de Lagrange, resultando em amplificações indesejadas na função interpolante.
  
##### Resumo do Comportamento Observado
- Centro do Intervalo: A interpolação é mais precisa e estável no centro do intervalo, onde a distribuição dos pontos é mais equilibrada e o polinômio interpolante segue de perto a função de Runge.
- Extremos do Intervalo: A interpolação sofre com grandes oscilações, especialmente quando $n$ aumenta. Isso é uma consequência direta do fenômeno de Runge, onde os polinômios de alta ordem oscilam fortemente nas extremidades do intervalo.

#### d) Pesquise na literatura e encontre uma alternativa para aprimorar a interpolação descartando o uso de Splines ou qualquer estratégia de interpolação por partes.

Na literatura esse comportamento oscilatório dos polínômios pode ser minimizado através da escolha de pontos não igualmente espaçados. Uma estratégia é baseada nos pontos de Chebyshev. Ela se mostra eficiente no controle do comportamento dos polinômios interpolantes para a função de Runge e garante a convergência. Para melhorar a interpolação da função de Runge utilizando pontos de Chebyshev, podemos seguir um processo similar ao anterior, mas calculando os pontos de interpolação de Chebyshev em vez de pontos igualmente espaçados.

Os pontos de Chebyshev são dados por:

$x_i = a\cos(\frac{2i+1}{2n}π)$, $i=0,1,2,…,n−1$.

Para o intervalo $[−5,5]$ eles são:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função de Runge
def funcao_runge(x):
    return 1 / (1 + x**2)

# Definindo os pontos de Chebyshev
def chebyshev_points(a, b, n):
    k = np.arange(n)
    x = np.cos((2*k + 1) * np.pi / (2*n))
    # Transformando para o intervalo [a, b]
    return 0.5 * (a + b) + 0.5 * (b - a) * x

# Calculando os polinômios de Lagrange
def lagrange_basicos(xi, x):
    n = len(xi)
    L = np.ones((n, len(x)))
    for i in range(n):
        for j in range(n):
            if i != j:
                L[i] *= (x - xi[j]) / (xi[i] - xi[j])
    return L

# Construindo o polinômio interpolante de Lagrange
def polinomio_interpolante_lagrange(xi, fi, x):
    L = lagrange_basicos(xi, x)
    P = np.dot(fi, L)
    return P

# Intervalo e número de pontos
a, b = -5, 5
pontos = [4, 6, 8, 12]
x_plot = np.linspace(a, b, 1000)

# Plotando a função original e os polinômios interpolantes para cada n usando pontos de Chebyshev
for n in pontos:
    xi = chebyshev_points(a, b, n)
    fi = funcao_runge(xi)
    P = polinomio_interpolante_lagrange(xi, fi, x_plot)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_plot, funcao_runge(x_plot), label='Função de Runge', color='black')
    plt.plot(x_plot, P, label=f'Polinômio de Lagrange (n={n})')
    label_text = '\n'.join([f'Ponto Cheb {i+1} = ({xi[i]:.2f}, {fi[i]:.2f})' for i in range(n)])
    plt.scatter(xi, fi, color='red', label=label_text)  # pontos de interpolação
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Polinômio de Lagrange Interpolante com Pontos de Chebyshev (n={n})')
    plt.grid(True)
    plt.show()
```
