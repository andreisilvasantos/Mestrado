### 1. Considere a função f(x) = sen(2x) + cos(3x)

##### a) Obtenha o polinômio de Taylor de grau 2 [P2(x)], e de grau 3 [P3(x)], que aproximem a funcão no ponto x = π;

```python
import numpy as np
import math
import matplotlib.pyplot as plt

a = math.pi

def fx(x):
    return math.sin(2 * x) + math.cos(3 * x) 

def fx1(x):
    return 2 * math.cos(2 * x) -3 * math.sin(3 * x) 

def fx2(x):
    return -4 * math.sin(2 * x) -9 * math.cos(3 * x) 

def fx3(x):
    return -8 * math.cos(2 * x) -27 * math.sin(3 * x) 

def pol_taylor_G2(x):
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2)

def pol_taylor_G3(x):
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2) + (fx3(a) * (x - a) ** 3) / math.factorial(3)
```

```python
x1 = a + 0.1
x2 = a - 0.5

print("Valores Verdadeiros")
print(fx(x1))
print(fx(x2))

print("\nValores estimados por P(2)")
print("x = pi + 0.1. Resultado ", pol_taylor_G2(x1))
print("x = pi - 0.5. Resultado ", pol_taylor_G2(x2))

print("\nErro absoluto de P(2)")
print(abs(pol_taylor_G2(x1) - fx(x1)))
print(abs(pol_taylor_G2(x2) - fx(x2)))

print("\nValores estimados por P(2)")
print("x = pi + 0.1. Resultado ", pol_taylor_G2(x1))
print("x = pi - 0.5. Resultado ", pol_taylor_G2(x2))
print("\nErro absoluto de P(3)")
print(abs(pol_taylor_G3(x1) - fx(x1)))
print(abs(pol_taylor_G3(x2) - fx(x2)))
```

#### c) Obtenha um limite superior para os polinômios de Taylor P2(x) e P3(x);

Os polinômios de Taylor fornecem uma aproximação local para funções diferenciáveis. O erro da aproximação é dado pelo termo residual, que pode ser estimado usando o Teorema de Lagrange. Neste problema, vamos determinar o limite superior para o erro dos polinômios de Taylor de segundo e terceiro graus para a função f(x) = sen(2x) + cos(3x), em torno do ponto x = π.

Desenvolvimento:

1. Polinômio de Taylor de Segundo Grau:

    O polinômio de Taylor de segundo grau para f(x) em x = π é dado por:

    P₂(x) = f(π) + f'(π)(x - π) + f''(π)(x - π)²/2

    Substituindo f(x), f'(x) e f''(x) com suas respectivas definições:

    P₂(x) = sen(2π) + 2cos(2π)(x - π) - 4sen(2π)(x - π)²/2

    P₂(x) = 0 + 2(-1)(x - π) - 4(0)(x - π)²/2

    P₂(x) = -2(x - π)

2. Limite Superior para o Erro:

    O erro do polinômio de Taylor de segundo grau é dado pelo termo residual:

    R₂(x) = f'''(ξ)(x - π)³/6

    Para encontrar o limite superior do erro, precisamos encontrar o valor máximo de |f'''(ξ)| no intervalo [π, x].

    |f'''(ξ)| = |6cos(2ξ)| ≤ 6

    Portanto, o limite superior para o erro do polinômio de Taylor de segundo grau é:

    |R₂(x)| ≤ 6|x - π|³/6 = |x - π|³


3. Polinômio de Taylor de Terceiro Grau:

    O polinômio de Taylor de terceiro grau para f(x) em x = π é dado por:

    P₃(x) = P₂(x) + f'''(π)(x - π)³/6

    Substituindo f'''(π) = 12cos(2π) = 12:

    P₃(x) = -2(x - π) + 12(x - π)³/6

    P₃(x) = -2(x - π) + 2(x - π)³

4. Limite Superior para o Erro:

    O erro do polinômio de Taylor de terceiro grau é dado pelo termo residual:

    R₃(x) = f''''(ξ)(x - π)⁴/24

    Para encontrar o limite superior do erro, precisamos encontrar o valor máximo de |f''''(ξ)| no intervalo [π, x].

    |f''''(ξ)| = |-12sen(2ξ)| ≤ 12

    Portanto, o limite superior para o erro do polinômio de Taylor de terceiro grau é:

    |R₃(x)| ≤ 12|x - π|⁴/24 = |x - π|⁴/2

Conclusão:

O limite superior para o erro do polinômio de Taylor de segundo grau para f(x) = sen(2x) + cos(3x) em x = π é |x - π|³. Já o limite superior para o erro do polinômio de Taylor de terceiro grau é |x - π|⁴/2. Isso significa que o polinômio de Taylor de terceiro grau fornece uma aproximação mais precisa para f(x) em torno de x = π do que o polinômio de Taylor de segundo grau.

### 2. Dada a função $f(x) = ln(x + 1)$, e os pontos x0 =0,0; x1 =0,6; x2=0,9 e x3=1,1. Construa os polinômios de interpolacão de Lagrange nos seguintes casos:

Para construir os polinômios de interpolação de Lagrange, utilizamos a seguinte fórmula geral para um polinômio $ P(x) $ de grau $ n $:

$P(x) = \sum_{i=0}^{n} y_i \cdot L_i(x)$

onde $ L_i(x) $ são os polinômios básicos de Lagrange definidos por:

$ L_i(x) = \prod_{\substack{0 \le j \le n \\ j \ne i}} \frac{x - x_j}{x_i - x_j} $

Os pontos dados são:
- $ x_0 = 0,0 $
- $ x_1 = 0,6 $
- $ x_2 = 0,9 $
- $ x_3 = 1,1 $

A função dada é $ f(x) = \ln(x + 1) $. Portanto, os valores de $ f(x) $ nos pontos dados são:
- $ f(x_0) = \ln(0,0 + 1) = \ln(1) = 0 $
- $ f(x_1) = \ln(0,6 + 1) = \ln(1,6) \approx 0,470 $
- $ f(x_2) = \ln(0,9 + 1) = \ln(1,9) \approx 0,641 $
- $ f(x_3) = \ln(1,1 + 1) = \ln(2,1) \approx 0,742 $

### a) Polinômio de grau 1, com os pontos $ x_0 $ e $ x_3 $

Para um polinômio de grau 1 utilizando $ x_0 $ e $ x_3 $:

$ P_1(x) = y_0 \cdot L_0(x) + y_3 \cdot L_3(x) $

onde:
$ L_0(x) = \frac{x - x_3}{x_0 - x_3} = \frac{x - 1,1}{0 - 1,1} = -\frac{x - 1,1}{1,1} $

$ L_3(x) = \frac{x - x_0}{x_3 - x_0} = \frac{x - 0}{1,1 - 0} = \frac{x}{1,1} $

Então:
$ P_1(x) = 0 \cdot L_0(x) + 0,742 \cdot L_3(x) = 0,742 \cdot \frac{x}{1,1} = \frac{0,742}{1,1}x $

$ P_1(x) = 0,6745x $

### b) Polinômio de grau 2, com os pontos $ x_0 $, $ x_1 $ e $ x_3 $

Para um polinômio de grau 2 utilizando $ x_0 $, $ x_1 $ e $ x_3 $:

$ P_2(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_3 \cdot L_3(x) $

onde:
$ L_0(x) = \frac{(x - x_1)(x - x_3)}{(x_0 - x_1)(x_0 - x_3)} = \frac{(x - 0,6)(x - 1,1)}{(0 - 0,6)(0 - 1,1)} = \frac{(x - 0,6)(x - 1,1)}{0,66} $

$ L_1(x) = \frac{(x - x_0)(x - x_3)}{(x_1 - x_0)(x_1 - x_3)} = \frac{(x - 0)(x - 1,1)}{(0,6 - 0)(0,6 - 1,1)} = \frac{x(x - 1,1)}{-0,3} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)}{(x_3 - x_0)(x_3 - x_1)} = \frac{(x - 0)(x - 0,6)}{(1,1 - 0)(1,1 - 0,6)} = \frac{x(x - 0,6)}{0,55} $

Os valores de $ y_i $ são:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_3 = 0,742 $

Substituindo, obtemos:
$ P_2(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:
$ P_2(x) = 0 + 0,470 \cdot \frac{x(x - 1,1)}{-0,3} + 0,742 \cdot \frac{x(x - 0,6)}{0,55} $

### c) Polinômio de grau 3, com os pontos $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $

Para um polinômio de grau 3 utilizando $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $:

$ P_3(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_2 \cdot L_2(x) + y_3 \cdot L_3(x) $

onde:
$ L_0(x) = \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)} $

$ L_1(x) = \frac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)} $

$ L_2(x) = \frac{(x - x_0)(x - x_1)(x - x_3)}{(x_2 - x_0)(x_2 - x_1)(x_2 - x_3)} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)(x - x_2)}{(x_3 - x_0)(x_3 - x_1)(x_3 - x_2)} $

Os valores de $ y_i $ são:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_2 = 0,641 $
- $ y_3 = 0,742 $

Substituindo, obtemos:
$ P_3(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:
$ P_3(x) = 0 + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

### Polinômios Básicos de Lagrange

Os polinômios básicos $ L_i(x) $ para $ P_3(x) $ são:

- $ L_1(x) = \frac{(x - 0)(x - 0,9)(x - 1,1)}{(0,6 - 0)(0,6 - 0,9)(0,6 - 1,1)} $
- $ L_2(x) = \frac{(x - 0)(x - 0,6)(x - 1,1)}{(0,9 - 0)(0,9 - 0,6)(0,9 - 1,1)} $
- $ L_3(x) = \frac{(x - 0)(x - 0,6)(x - 0,9)}{(1,1 - 0)(1,1 - 0,6)(1,1 - 0,9)} $

Substituindo os valores específicos:
- $ L_1(x) = \frac{(x)(x - 0,9)(x - 1,1)}{0,6 \cdot (-0,3) \cdot (-0,5)} = \frac{(x)(x - 0,9)(x - 1,1)}{0,09} $
- $ L_2(x) = \frac{(x)(x - 0,6)(x - 1,1)}{0,9 \cdot 0,3 \cdot (-0,2)} = \frac{(x)(x - 0,6)(x - 1,1)}{-0,054} $
- $ L_3(x) = \frac{(x)(x - 0,6)(x - 0,9)}{1,1 \cdot 0,5 \cdot 0,2} = \frac{(x)(x - 0,6)(x - 0,9)}{0,11} $

Portanto:
$ P_3(x) = 0 + 0,470 \cdot \frac{(x)(x - 0,9)(x - 1,1)}{0,09} + 0,641 \cdot \frac{(x)(x - 0,6)(x - 1,1)}{-0,054} + 0,742 \cdot \frac{(x)(x - 0,6)(x - 0,9)}{0,11} $

#### Construa um gráfico com a função f(x), os pontos de interpolação e os polinômios P1, P2 e P3. Utilize os polinômios de interpolacão para avaliar a função nos pontos x = 0,3 e x = 0,75. Calcule o desvio relativo com relação ao valor verdadeiro em cada caso. Comente os resultados obtidos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função original
def f(x):
    return np.log(x + 1)

# Pontos dados
x0, x1, x2, x3 = 0.0, 0.6, 0.9, 1.1
y0, y1, y2, y3 = f(x0), f(x1), f(x2), f(x3)

# Polinômio de grau 1
def P1(x):
    return 0.6745 * x

# Polinômio de grau 2
def P2(x):
    return 0.470 * x * (x - 1.1) / - 0.3 + 0.742 * x * (x - 0.6) / 0.55

# Polinômio de grau 3
def P3(x):
    return (0.470 * x * (x - 0.9) * (x - 1.1) / 0.09 +
            0.641 * x * (x - 0.6) * (x - 1.1) / - 0.054 +
            0.742 * x * (x - 0.6) * (x - 0.9) / 0.11)

# Valores de x para plotagem
x_values = np.linspace(-0.1, 1.2, 400)

# Valores de f(x)
y_values = f(x_values)

# Valores dos polinômios
P1_values = P1(x_values)
P2_values = P2(x_values)
P3_values = P3(x_values)

# Pontos de interpolação
x_points = [x0, x1, x2, x3]
y_points = [y0, y1, y2, y3]

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = ln(x + 1)', color='black')
plt.plot(x_values, P1_values, label='P1(x)', linestyle='--')
plt.plot(x_values, P2_values, label='P2(x)', linestyle='--')
plt.plot(x_values, P3_values, label='P3(x)', linestyle='--')
plt.scatter(x_points, y_points, label='Dados', color='red', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolação de Lagrange')
plt.grid(True)
plt.show()
```

```python
# Avaliação nos pontos x = 0.3 e x = 0.75
x_test = [0.3, 0.75]
f_values = f(np.array(x_test))
P1_test = P1(np.array(x_test))
P2_test = P2(np.array(x_test))
P3_test = P3(np.array(x_test))

# Desvios relativos
desvio_relativo_P1 = np.abs((P1_test - f_values) / f_values) * 100
desvio_relativo_P2 = np.abs((P2_test - f_values) / f_values) * 100
desvio_relativo_P3 = np.abs((P3_test - f_values) / f_values) * 100

# Resultados
print("X de teste:", x_test)
print("Valores verdadeiros de f(x):", f_values)
print("Polinôminio de Grau 1:", P1_test)
print("Polinôminio de Grau 2:", P2_test)
print("Polinôminio de Grau 3:", P3_test)
print("Desvio relativo P1 (%):", np.round(desvio_relativo_P1, 2))
print("Desvio relativo P2 (%):", np.round(desvio_relativo_P2, 2))
print("Desvio relativo P3 (%):", np.round(desvio_relativo_P3, 2))
```

Comentários sobre os resultados:

- Polinômio de grau 1 (P1​ (x)): O desvio relativo é significativamente maior em ambos os pontos (22.87% para x=0.3 e 9.60% para x=0.75). Isso mostra que o polinômio de grau 1 não é suficiente para uma boa aproximação da função f(x), especialmente em pontos distantes dos nós de interpolação. 

- Polinômio de grau 2 (P2 ​ (x)): O desvio relativo é muito menor em comparação ao polinômio de grau 1 (2.96% para x=0.3 e 0.61% para x=0.75). Isso indica que o polinômio de grau 2 proporciona uma aproximação significativamente melhor da função f(x). 

- Polinômio de grau 3 (P3 (x)): O desvio relativo é extremamente pequeno (0.74% para x=0.3 e 0.04% para x=0.75), o que sugere que o polinômio de grau 3 é capaz de capturar muito bem a forma da função f(x), fornecendo uma aproximação quase exata nos pontos avaliados.
