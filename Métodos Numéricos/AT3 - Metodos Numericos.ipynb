### 1. Considere a funo f(x) = sen(2x) + cos(3x)

##### a) Obtenha o polinmio de Taylor de grau 2 [P2(x)], e de grau 3 [P3(x)], que aproximem a funco no ponto x = ;

f(x)=2cos(2x)3sen(3x)

f(x)=-4sen(2x)9cos(3x)

f(x)=-8cos(2x)+27sen(3x)



```python
import numpy as np
import math
import matplotlib.pyplot as plt

def fx(x):
    return math.sin(2 * x) + math.cos(3 * x) 

def fx1(x):
    return 2 * math.cos(2 * x) -3 * math.sin(3 * x) 

def fx2(x):
    return -4 * math.sin(2 * x) -9 * math.cos(3 * x) 

def fx3(x):
    return -8 * math.cos(2 * x) -27 * math.sin(3 * x) 

def pol_taylor_G2(x):
    return fx(math.pi) + fx1(math.pi) * (x - math.pi) + (fx2(math.pi) * ( x - math.pi) ** 2) / math.factorial(2)

def pol_taylor_G3(math.pi, x):
    return fx(math.pi) + fx1(math.pi) * (x - math.pi) + (fx2(math.pi) * ( x - math.pi) ** 2) / math.factorial(2) + (fx3(math.pi) * (x - math.pi) ** 3) / math.factorial(3)
```
