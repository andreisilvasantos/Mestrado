### 1. Considere a funo $f(x) = sen(2x) + cos(3x)$

#### Grfico de $f(x)$


1. Polinmio de Taylor de Segundo Grau:

    O polinmio de Taylor de segundo grau para $f(x)$ em $x = $  dado por:

    $P(x) = f() + f'()(x - ) + f''()(x - )/2$

    Substituindo $f(x), f'(x) e f''(x)$ com suas respectivas definies:

    $P(x) = sen(2) + 2cos(2)(x - ) - 4sen(2)(x - )/2$

    $P(x) = 0 + 2(-1)(x - ) - 4(0)(x - )/2$

    $P(x) = -2(x - )$





2. Polinmio de Taylor de Terceiro Grau:

    O polinmio de Taylor de terceiro grau para f(x) em x =   dado por:

    $P(x) = P(x) + f'''()(x - )/6$

    Substituindo $f'''() = 12cos(2) = 12$:

    $P(x) = -2(x - ) + 12(x - )/6$

    $P(x) = -2(x - ) + 2(x - )$



##### Cdigo em Python

```python
import numpy as np
import math
import matplotlib.pyplot as plt

a = np.pi

def fx(x):
    return np.sin(2 * x) + np.cos(3 * x) 

def fx1(x):
    return 2 * np.cos(2 * x) -3 * np.sin(3 * x) 

def fx2(x):
    return -4 * np.sin(2 * x) -9 * np.cos(3 * x) 

def fx3(x):
    return -8 * np.cos(2 * x) -27 * np.sin(3 * x) 

def pol_taylor_G2(x):
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2)

def pol_taylor_G3(x):
    #return -2 * (x - np.pi) + 2 * (x - np.pi) ** 3
    return fx(a) + fx1(a) * (x - a) + (fx2(a) * ( x - a) ** 2) / math.factorial(2) + (fx3(a) * (x - a) ** 3) / math.factorial(3)
```

### 2. Dada a funo $f(x) = ln(x + 1)$, e os pontos $x_0 =0,0; $ $x_1 =0,6; $ $x_2=0,9$ e $x_3=1,1$. Construa os polinmios de interpolaco de Lagrange nos seguintes casos:

Para construir os polinmios de interpolao de Lagrange, utilizamos a seguinte frmula geral para um polinmio $ P(x) $ de grau $ n $:

$P(x) = \sum_{i=0}^{n} f(x_i) \cdot L_i(x)$

onde $ L_i(x) $ so os polinmios bsicos de Lagrange definidos por:

$ L_i(x) = \prod_{\substack{0 \le j \le n \\ j \ne i}} \frac{x - x_j}{x_i - x_j} $

Os pontos dados so:
- $ x_0 = 0,0 $
- $ x_1 = 0,6 $
- $ x_2 = 0,9 $
- $ x_3 = 1,1 $

A funo dada  $ f(x) = \ln(x + 1) $. Portanto, os valores de $ f(x) $ nos pontos dados so:
- $ f(x_0) = \ln(0,0 + 1) = \ln(1) = 0 $
- $ f(x_1) = \ln(0,6 + 1) = \ln(1,6) \approx 0,470 $
- $ f(x_2) = \ln(0,9 + 1) = \ln(1,9) \approx 0,641 $
- $ f(x_3) = \ln(1,1 + 1) = \ln(2,1) \approx 0,742 $

#### a) Polinmio de grau 1, com os pontos $ x_0 $ e $ x_3 $

Para um polinmio de grau 1 utilizando $ x_0 $ e $ x_3 $:

$ P_1(x) = y_0 \cdot L_0(x) + y_3 \cdot L_3(x) $

onde:

$ L_0(x) = \frac{x - x_3}{x_0 - x_3} = \frac{x - 1,1}{0 - 1,1} = -\frac{x - 1,1}{1,1} $

$ L_3(x) = \frac{x - x_0}{x_3 - x_0} = \frac{x - 0}{1,1 - 0} = \frac{x}{1,1} $

Ento:

$ P_1(x) = 0 \cdot L_0(x) + 0,742 \cdot L_3(x)$

$ P_1(x) = 0,742 \cdot \frac{x}{1,1} = \frac{0,742}{1,1}x $

$ P_1(x) = 0,6745x $

### b) Polinmio de grau 2, com os pontos $ x_0 $, $ x_1 $ e $ x_3 $

Para um polinmio de grau 2 utilizando $ x_0 $, $ x_1 $ e $ x_3 $:

$ P_2(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_3 \cdot L_3(x) $

onde:

$ L_0(x) = \frac{(x - x_1)(x - x_3)}{(x_0 - x_1)(x_0 - x_3)} = \frac{(x - 0,6)(x - 1,1)}{(0 - 0,6)(0 - 1,1)} = \frac{(x - 0,6)(x - 1,1)}{0,66} $

$ L_1(x) = \frac{(x - x_0)(x - x_3)}{(x_1 - x_0)(x_1 - x_3)} = \frac{(x - 0)(x - 1,1)}{(0,6 - 0)(0,6 - 1,1)} = \frac{x(x - 1,1)}{-0,3} $

$ L_3(x) = \frac{(x - x_0)(x - x_1)}{(x_3 - x_0)(x_3 - x_1)} = \frac{(x - 0)(x - 0,6)}{(1,1 - 0)(1,1 - 0,6)} = \frac{x(x - 0,6)}{0,55} $

Os valores de $ y_i $ so:
- $ y_0 = 0 $
- $ y_1 = 0,470 $
- $ y_3 = 0,742 $

Substituindo, obtemos:

$ P_2(x) = 0 \cdot L_0(x) + 0,470 \cdot L_1(x) + 0,742 \cdot L_3(x) $

Substituindo $ L_i(x) $:

$ P_2(x) = 0 + 0,470 \cdot \frac{x(x - 1,1)}{-0,3} + 0,742 \cdot \frac{x(x - 0,6)}{0,55} $

### c) Polinmio de grau 3, com os pontos $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $

Para um polinmio de grau 3 utilizando $ x_0 $, $ x_1 $, $ x_2 $ e $ x_3 $:

$ P_3(x) = y_0 \cdot L_0(x) + y_1 \cdot L_1(x) + y_2 \cdot L_2(x) + y_3 \cdot L_3(x) $

onde:
$ L_0(x) = \frac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)} $

$ L_1(x) = \frac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)} $

$ L_2(x) = \frac{(x - x_0)(x - x_1)(x - x_3)}{(x_2 - x_0)(x_2 - x_1)(x_2 - x_3)} $

Os valores de $ y_i $ so:

Substituindo $ L_i(x) $:
$ P_3(x) = 0 + 0,470 \cdot L_1(x) + 0,641 \cdot L_2(x) + 0,742 \cdot L_3(x) $

### Construa um grfico com a funo f(x), os pontos de interpolao e os polinmios $P_1$, $P_2$ e $P_3$.
```python
import numpy as np
import matplotlib.pyplot as plt

# Funo original
def f(x):
    return np.log(x + 1)

# Pontos dados
x0, x1, x2, x3 = 0.0, 0.6, 0.9, 1.1
y0, y1, y2, y3 = f(x0), f(x1), f(x2), f(x3)

# Polinmio de grau 1
def P1(x):
    return 0.6745 * x

# Polinmio de grau 2
def P2(x):
    return 0.470 * x * (x - 1.1) / - 0.3 + 0.742 * x * (x - 0.6) / 0.55

# Polinmio de grau 3
def P3(x):
    return (0.470 * x * (x - 0.9) * (x - 1.1) / 0.09 +
            0.641 * x * (x - 0.6) * (x - 1.1) / - 0.054 +
            0.742 * x * (x - 0.6) * (x - 0.9) / 0.11)

# Valores de x para plotagem
x_valores = np.linspace(-0.1, 1.2, 400)

# Valores de f(x)
y_valores = f(x_valores)

# Valores dos polinmios
P1_y = P1(x_valores)
P2_y = P2(x_valores)
P3_y = P3(x_valores)

# Pontos de interpolao
x_pontos = [x0, x1, x2, x3]
y_pontos = [y0, y1, y2, y3]

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(x_valores, y_valores, label='f(x) = ln(x + 1)', color='black')
plt.plot(x_valores, P1_y, label='P1(x)', linestyle='--')
plt.plot(x_valores, P2_y, label='P2(x)', linestyle='--')
plt.plot(x_valores, P3_y, label='P3(x)', linestyle='--')
plt.scatter(x_pontos, y_pontos, label='Dados', color='red', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolao de Lagrange')
plt.grid(True)
plt.show()
```

### 3. A funo de Runge  definida como f(x) = 1/(1 + x).

#### a) Desejamos construir o polinmio interpolante de Lagrange no intervalo [5, 5] com pontos equidistantes para n = 4, 6, 8 e 12.

O polinmio interpolante de Lagrange $P(x)$  uma combinao linear dos valores da funo $f$ nos pontos de interpolao, ponderada pelos polinmios base de Lagrange $l_i(x)$. A frmula geral do polinmio de Lagrange para n pontos $(x_0, f(x_0)) , (x_1, f(x_1)),,(x_{n-1},f(x_{n-1}))$  dada por: 

$P(x) = \sum_{i=0}^{n-1} y_i \cdot L_i(x)$

onde $ L_i(x) $ so os polinmios bsicos de Lagrange definidos por:

$ L_i(x) = \prod_{\substack{0 \le j \le n \\ j \ne i}} \frac{x - x_j}{x_i - x_j} $

Esta frmula significa que cada $l_i(x)$  um polinmio que  igual a 1 no ponto $x_i$ e igual a 0 nos demais pontos $x_j$ (para $j \ne i$). Portanto, o polinmio de Lagrange $P(x)$ passa exatamente pelos pontos de interpolao fornecidos. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a funo de Runge
def funcao_runge(x):
    return 1 / (1 + x**2)

# Definindo os pontos equidistantes e valores da funo
def pontos_equidistantes(a, b, n):
    return np.linspace(a, b, n)

# Calculando os polinmios de Lagrange
def lagrange_basicos(xi, x):
    n = len(xi)
    L = np.ones((n, len(x)))
    for i in range(n):
        for j in range(n):
            if i != j:
                L[i] *= (x - xi[j]) / (xi[i] - xi[j])
    return L

# Construindo o polinmio interpolante de Lagrange
def polinomio_interpolante_lagrange(xi, fi, x):
    L = lagrange_basicos(xi, x)
    P = np.dot(fi, L)
    return P
```

#### b) Em cada caso construa um grfico contendo a funo de Runge, os pontos de interpolao e o Polinmio de Lagrange.

```python
# Intervalo e nmero de pontos
a, b = -5, 5
pontos = [4, 6, 8, 12]
x_plot = np.linspace(a, b, 1000)

# Plotando a funo original e os polinmios interpolantes para cada n
for n in pontos:
    xi = pontos_equidistantes(a, b, n)
    fi = funcao_runge(xi)
    P = polinomio_interpolante_lagrange(xi, fi, x_plot)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_plot, funcao_runge(x_plot), label='Funo de Runge', color='black')
    plt.plot(x_plot, P, label=f'Polinmio de Lagrange (n={n})')
    plt.scatter(xi, fi, color='red')  # pontos de interpolao
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Polinmio de Lagrange Interpolante (n={n})')
    plt.grid(True)
    plt.show()
```

#### c) A partir dos grficos comente sobre a acurcia da interpolao no centro do intervalo, e nos extremos. Explique o comportamento observado.

Os grficos dos polinmios interpolantes de Lagrange para a funo de Runge com $n=4,6, 8, 12$ pontos igualmente espaados mostram alguns comportamentos tpicos da interpolao polinomial, especialmente em relao  acurcia no centro do intervalo e nos extremos. Vamos analisar essas caractersticas mais detalhadamente:

##### Acurcia no Centro do Intervalo

No centro do intervalo $[5,5]$, os polinmios interpolantes de Lagrange geralmente fornecem uma boa aproximao da funo de Runge. Para $n=4, 6 ,8 e 12$ os polinmios interpolantes seguem bem a forma da funo original, com a acurcia melhorando  medida que aumentamos o nmero de pontos. Este comportamento ocorre porque:

- Menores Erros: A interpolao polinomial tende a ser mais precisa prximo aos pontos de interpolao. No centro do intervalo, os pontos de interpolao esto mais prximos uns dos outros, reduzindo os erros de interpolao.
- Menor Oscilao: A oscilao (fenmeno de Runge)  menos pronunciada no centro do intervalo, onde os pontos so mais uniformemente distribudos em relao ao ponto avaliado.
  
##### Acurcia nos Extremos do Intervalo

Nos extremos do intervalo $[5,5]$, a acurcia da interpolao polinomial de Lagrange diminui significativamente, especialmente com o aumento do nmero de pontos de interpolao. Observamos oscilaes pronunciadas nos grficos para $n=8 e 12$. Este comportamento pode ser explicado por:

- Fenmeno de Runge: Este fenmeno descreve a tendncia dos polinmios interpolantes de alta ordem de oscilar fortemente nas extremidades do intervalo de interpolao.  medida que aumentamos o grau do polinmio, a oscilao nas extremidades se intensifica.
- Distribuio dos Pontos: Em um conjunto de pontos igualmente espaados, a densidade de pontos de interpolao  menor nas extremidades do intervalo. Isso pode levar a grandes valores nos denominadores dos polinmios de base de Lagrange, resultando em amplificaes indesejadas na funo interpolante.
  
##### Resumo do Comportamento Observado
- Centro do Intervalo: A interpolao  mais precisa e estvel no centro do intervalo, onde a distribuio dos pontos  mais equilibrada e o polinmio interpolante segue de perto a funo de Runge.
- Extremos do Intervalo: A interpolao sofre com grandes oscilaes, especialmente quando $n$ aumenta. Isso  uma consequncia direta do fenmeno de Runge, onde os polinmios de alta ordem oscilam fortemente nas extremidades do intervalo.

Para melhorar a interpolao da funo de Runge utilizando pontos de Chebyshev, podemos seguir um processo similar ao anterior, mas calculando os pontos de interpolao de Chebyshev em vez de pontos igualmente espaados.

Os pontos de Chebyshev so dados por:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a funo de Runge
def funcao_runge(x):
    return 1 / (1 + x**2)

# Definindo os pontos de Chebyshev
def chebyshev_points(a, b, n):
    k = np.arange(n)
    x = np.cos((2*k + 1) * np.pi / (2*n))
    # Transformando para o intervalo [a, b]
    return 0.5 * (a + b) + 0.5 * (b - a) * x

# Calculando os polinmios de Lagrange
def lagrange_basicos(xi, x):
    n = len(xi)
    L = np.ones((n, len(x)))
    for i in range(n):
        for j in range(n):
            if i != j:
                L[i] *= (x - xi[j]) / (xi[i] - xi[j])
    return L

# Construindo o polinmio interpolante de Lagrange
def polinomio_interpolante_lagrange(xi, fi, x):
    L = lagrange_basicos(xi, x)
    P = np.dot(fi, L)
    return P

# Intervalo e nmero de pontos
a, b = -5, 5
pontos = [4, 6, 8, 12]
x_plot = np.linspace(a, b, 1000)

# Plotando a funo original e os polinmios interpolantes para cada n usando pontos de Chebyshev
for n in pontos:
    xi = chebyshev_points(a, b, n)
    fi = funcao_runge(xi)
    P = polinomio_interpolante_lagrange(xi, fi, x_plot)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_plot, funcao_runge(x_plot), label='Funo de Runge', color='black')
    plt.plot(x_plot, P, label=f'Polinmio de Lagrange (n={n})')
    label_text = '\n'.join([f'Ponto Cheb {i+1} = ({xi[i]:.2f}, {fi[i]:.2f})' for i in range(n)])
    plt.scatter(xi, fi, color='red', label=label_text)  # pontos de interpolao
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f'Polinmio de Lagrange Interpolante com Pontos de Chebyshev (n={n})')
    plt.grid(True)
    plt.show()
```
