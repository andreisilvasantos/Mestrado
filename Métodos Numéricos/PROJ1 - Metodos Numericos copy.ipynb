### Gerando as 3 Matrizes de Hilbert do Exerccio

```python
import numpy as np
import time 
import matplotlib.pyplot as plt

def matriz_hilbert(n):
    matrizH = np.zeros((n, n))
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            matrizH[i-1, j-1] = 1 / (i + j - 1)
    return matrizH

def print_matriz(matriz):
    for row in matriz:
        print("|", end=" ")
        for elem in row:
            print("{:.3f}".format(elem), end=" ")
        print("|")

n = [5, 9, 15]
metodos = ['Gauss com Substituio Regressiva','Gauss com Pivoteamento Parcial','Gauss com Pivoteamento Parcial com Escala','Gauss com Pivoteamento total','Jacobi','Gauss-Seidel','Sobre-Relaxamento']
matrizes = [matriz_hilbert(i) for i in n] # Gerando as matrizes de Hilbert
B = [np.sum(matrizes[i], axis=1) for i in range(len(matrizes))] # Gerando os vetores B

metodos_vetor_solucao = [[np.zeros(len(matrizes[i])) for i in range(len(n))] for linha in metodos]
metodos_tempo_execucao = np.zeros((7, 3))
vetores_solucao = [np.ones(i) for i in n]

for j in range(0, len(matrizes)):
    print("Matriz de Hilbert de Ordem", n[j], ":\n")
    print_matriz(matrizes[j])
    print("\n")
```

#### a) Eliminao de Gauss com substituio regressiva

```python
# O indice [n, -1] da matriz aumentada  a coluna que representa o vetor independente bi
# Vetor solucao xi  instanciado com o nmero de linhas da matriz e valores iguais a zero

def eliminacao_Gauss_SR(matriz):
    n = len(matriz) - 1 ## Removendo o b

    # Fase eliminao de Gauss
    for k in range(n):
        for i in range(k + 1, n + 1):
            fator = matriz[i][k] / matriz[k][k]
            for j in range(k + 1, n + 1):
                matriz[i][j] -= fator * matriz[k][j]
            matriz[i][n+1] -= fator * matriz[k][n+1]

    # Fase substituio regressiva
    vetor_solucao = np.zeros(n+1)
    vetor_solucao[n] = matriz[n][n+1] / matriz[n][n]
    for i in range(n - 1, -1, -1):
        soma = matriz[i][n+1]
        for j in range(i + 1, n+1):
            soma -= matriz[i][j] * vetor_solucao[j]
        vetor_solucao[i] = soma / matriz[i][i]

    return vetor_solucao

for j in range(0, len(matrizes)):
    inicio = time.perf_counter() 
    matriz_aumentada = np.column_stack((matrizes[j], B[j]))

    print(f"\nMatriz Aumentada com n={len(matriz_aumentada)}:")
    print_matriz(matriz_aumentada)

    metodos_vetor_solucao[0][j] = eliminacao_Gauss_SR(matriz_aumentada)
    metodos_tempo_execucao[0][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milisengundos

    print("\nVetor Soluo: ", metodos_vetor_solucao[0][j])

    print("\nMatriz Resultante:")
    print_matriz(matriz_aumentada)

```

#### b) Eliminao de Gauss com Pivoteamento Parcial

```python
def pivoteamento_parcial(matriz_aumentada):
    # Crie uma cpia da matriz aumentada para evitar a modificao da original
    matriz_pivotada = [linha.copy() for linha in matriz_aumentada]
    n = len(matriz_pivotada)
    for i in range(n):
        indice_maximo = i
        ## Encontrando o indice do elemento de maior valor absoluto
        for j in range(i+1, n):
            if abs(matriz_pivotada[j][i]) > abs(matriz_pivotada[indice_maximo][i]):
                indice_maximo = j
        
        # Troca a linha atual pela linha com o maior elemento
        matriz_pivotada[i], matriz_pivotada[indice_maximo] = matriz_pivotada[indice_maximo], matriz_pivotada[i]

    return matriz_pivotada

for j in range(0, len(matrizes)):
    inicio = time.perf_counter() 
    matriz_aumentada = np.column_stack((matrizes[j], B[j]))
    
    # Fase de Pivoteamento
    matriz_pivotada = pivoteamento_parcial(matriz_aumentada)
    print(f"\nMatriz Pivotada com n={len(matrizes[j])}:")
    print_matriz(matriz_pivotada)

    # Fase de Eliminaco de Gauss
    metodos_vetor_solucao[1][j] = eliminacao_Gauss_SR(matriz_pivotada)
    metodos_tempo_execucao[1][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milisengundos

    print("\nVetor Soluo: ", metodos_vetor_solucao[1][j])

    print("\nMatriz Resultante:")
    print_matriz(matriz_pivotada)
```

#### c) Eliminao de Gauss com Pivoteamento Parcial com Escala

```python
def pivoteamento_parcial_escala(matriz_aumentada):
    # Crie uma cpia da matriz aumentada para evitar a modificao da original
    matriz_pivotada = np.array(matriz_aumentada)
    n = len(matriz_pivotada)
    escalas = np.max(np.abs(matriz_pivotada[:][:]), axis=1)  # Calculando os mximos de cada linha

    for i in range(n):
        # Encontrando o indice do maior fator pivo / escala
        indice_maximo = i + np.argmax(np.abs(matriz_pivotada[i:, i]) / escalas[i:])
        
        # Troca da linha atual pela do pivo selecionado para minimizar efeitos de erros de arrendodamento
        matriz_pivotada[[i, indice_maximo]] = matriz_pivotada[[indice_maximo, i]]

    return matriz_pivotada

for j in range(0, len(matrizes)):
    inicio = time.perf_counter() 
    matriz_aumentada = np.column_stack((matrizes[j], B[j]))

    ## Fase de Pivoteamento Parcial por Escala
    matriz_pivotada = pivoteamento_parcial_escala(matriz_aumentada)
    print(f"\nMatriz Pivotada com n={len(matrizes[j])}:")
    print_matriz(matriz_pivotada)

    # Fase de Eliminaao de Gauss
    metodos_vetor_solucao[2][j] = eliminacao_Gauss_SR(matriz_pivotada)
    metodos_tempo_execucao[2][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milisengundos

    print("\nVetor Soluo: ", metodos_vetor_solucao[2][j])
    print("\nMatriz Resultante:")
    print_matriz(matriz_pivotada)
```

#### d) Eliminao de Gauss com Pivoteamento Total

```python
import numpy as np

def pivoteamento_total(matriz_aumentada):
    # Crie uma cpia da matriz aumentada para evitar a modificao da original
    matriz_pivotada = [linha.copy() for linha in matriz_aumentada]
    n = len(matriz_pivotada)
    
    for i in range(n):
        indice_maximo_linha, indice_maximo_coluna = i, i
        for j in range(i, n):
            for k in range(i, n):
                ## Encontrando o melhor pivo na linha e na coluna atual
                if abs(matriz_pivotada[j][k]) > abs(matriz_pivotada[indice_maximo_linha][indice_maximo_coluna]):
                    indice_maximo_linha, indice_maximo_coluna = j, k
        
        # Troca de linhas na matriz aumentada
        matriz_pivotada[i], matriz_pivotada[indice_maximo_linha] = matriz_pivotada[indice_maximo_linha], matriz_pivotada[i]

        # Troca de colunas na matriz aumentada
        for j in range(n):
            matriz_pivotada[j][i], matriz_pivotada[j][indice_maximo_coluna] = matriz_pivotada[j][indice_maximo_coluna], matriz_pivotada[j][i]

    return matriz_pivotada

for j in range(0, len(matrizes)):
    inicio = time.perf_counter() 
    matriz_aumentada = np.column_stack((matrizes[j], B[j]))
    # Fase do Pivoteamento Total
    matriz_pivotada = pivoteamento_total(matriz_aumentada)
    print("\nMatriz Pivotada:")
    print_matriz(matriz_pivotada)

    # Fase da eliminao de Gauss
    metodos_vetor_solucao[3][j] = eliminacao_Gauss_SR(matriz_pivotada)
    metodos_tempo_execucao[3][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milisengundos

    print("\nVetor Soluo:", metodos_vetor_solucao[3][j])
```

####  e) Jacobi

```python
import numpy as np

def jacobi(A, b, x0, tol=1e-4, max_iter=1000):
    # Inicializao do vetor de erro
    erro = np.zeros(len(x0))
    iter_count = 0

    # Loop principal de iterao
    for n in range(max_iter):
        # Clculo do novo vetor de soluo
        x_novo = np.array(x0)
        iter_count += 1

        for i in range(len(x0)):
            soma = 0
            for j in range(0, i):
                soma += A[i, j] * x0[j]
            for j in range(i+1, len(x0)):
                soma += A[i, j] * x0[j]
            x_novo[i] = (b[i] - soma) / A[i, i]

        # Clculo do erro
        erro = np.abs(x_novo - x0)

        # Verificao da convergncia
        if np.max(erro) < tol:
            return x_novo, iter_count  # Retorna a soluo aproximada quando convergir

    # Se no convergiu dentro do nmero mximo de iteraes
    # retorna vetor de infinitos(maior float representavel)
    return np.full(len(x0), np.inf), iter_count

# Exemplo de uso
for j in range(len(matrizes)):
    inicio = time.perf_counter() 
    matriz_atual = matrizes[j]
    x0 = [matriz_atual[j][-1] / matriz_atual[j][j] for j in range(len(matriz_atual))]

    print("\nMatriz Original:")
    print_matriz(matrizes[j])
    
    metodos_vetor_solucao[4][j], iter_count = jacobi(matrizes[j], B[j], x0)
    metodos_tempo_execucao[4][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milissegundos 

    if np.isinf(metodos_vetor_solucao[4][j]).any:
        print("\nO Mtodo de Jacobi no convergiu para uma soluo do Sistema.")
    else:
        print("\nVetor Soluo:", metodos_vetor_solucao[4][j])
        print(f"\nQuantidades de Iteraes necessrias: {iter_count}")
```

#### f) Gauss-Seidel

```python
import numpy as np

def metodo_gauss_seidel(A, b, x0, tol=1e-4, max_iter=1000):

  # Inicializao do vetor de erro
  erro = np.ones(len(x0))
  iter_count = 0

  # Loop principal de iterao
  for n in range(max_iter):
    iter_count += 1
    x_novo = np.array(x0)# Clculo do novo vetor de soluo
    for i in range(len(x0)):
      soma = 0
      for j in range(0, i):
        soma += A[i, j] * x_novo[j] # Usando o xi atualizado
      for j in range(i+1, len(x0)):
        soma += A[i, j] * x0[j]
      x_novo[i] = (b[i] - soma) / A[i, i]

    # Clculo do erro
    erro = np.abs(x_novo - x0)

    # Verificao da convergncia
    if np.max(erro) < tol:
      break

    # Atualizao do vetor de soluo
    x0 = x_novo

  # Retorno do vetor de soluo aproximado
  return x_novo, iter_count

# Exemplo de uso

for j in range(0, len(matrizes)):
    inicio = time.perf_counter() 
    matriz_aumentada = np.column_stack((matrizes[j], B[j]))
    x0 = np.zeros_like(B[j])  # Estimativa inicial

    print("\nMatriz Aumentada:")
    print_matriz(matriz_aumentada)

    metodos_vetor_solucao[5][j], iter_count = metodo_gauss_seidel(matrizes[j], B[j], x0)
    metodos_tempo_execucao[5][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milisengundos

    print("\nVetor Soluo:",  metodos_vetor_solucao[5][j])
    print(f"\nQuantidades de Iteraes necessrias: {iter_count}")
```

#### g) Gauss-Seidel com Sobre-relaxamento

```python
import numpy as np

def gauss_seidel_relaxamento(A, b, x0, omega, tol=1e-4, max_iter=1000):
    # Inicializao do vetor de erro
  erro = np.ones(len(x0))
  iter_count = 0

  # Loop principal de iterao
  for n in range(max_iter):
    iter_count += 1
    # Clculo do novo vetor de soluo
    x_novo = np.array(x0)
    for i in range(len(x0)):
      soma = 0
      for j in range(0, i):
        soma += A[i, j] * x_novo[j] # Usando o xi atualizado
      for j in range(i+1, len(x0)):
        soma += A[i, j] * x0[j]
      x_novo[i] = (1 - omega) * x0[i] + omega * (b[i] - soma) / A[i, i]

    # Clculo do erro
    erro = np.abs(x_novo - x0)

    # Verificao da convergncia
    if np.max(erro) < tol:
      break

    # Atualizao do vetor de soluo
    x0 = x_novo

  # Retorno do vetor de soluo aproximado
  return x_novo, iter_count

#Exemplo de uso

for j in range(0, len(matrizes)):
    inicio = time.perf_counter() 
    matriz_aumentada = np.column_stack((matrizes[j], B[j]))

    x0 = np.zeros_like(B[j])  # Estimativa inicial
    omega = 1.2  # Fator de relaxamento

    print("\nMatriz Aumentada:")
    print_matriz(matriz_aumentada)

    metodos_vetor_solucao[6][j], iter_count = gauss_seidel_relaxamento(matrizes[j], B[j], x0, omega)
    metodos_tempo_execucao[6][j] = (time.perf_counter()  - inicio) * 1000 # Convertendo para milisengundos

    print("\nVetor Soluo:",  metodos_vetor_solucao[6][j])
    print(f"\nQuantidades de Iteraes necessrias: {iter_count}")
```

### Mal Condicionamento das Matrizes de Hilbert
#### O que  mal condicionamento?

O condicionamento de uma matriz mede a sua sensibilidade a pequenas perturbaes nos dados. Uma matriz bem condicionada  pouco afetada por pequenas mudanas nos seus elementos, enquanto uma matriz mal condicionada  altamente sensvel a essas mudanas. No contexto de sistemas lineares, o mal condicionamento significa que pequenas alteraes nos termos independentes do sistema podem levar a grandes mudanas na soluo. Isso torna difcil a resoluo precisa do sistema por mtodos numricos, pois os erros de arredondamento introduzidos pelas operaes computacionais podem ser amplificados significativamente.

#### Por que as matrizes de Hilbert so mal condicionadas?

Existem diversas razes para o mal condicionamento das matrizes de Hilbert:

- Elevado nmero de condies: O valor de H[i, j] tende a zero para i e j grandes, o que significa que a matriz possui um grande nmero de condies prximas de zero. As condies so os autovalores da matriz inversa, e um grande nmero de condies prximas de zero indica que a matriz inversa  mal condicionada.
- Distribuio de autovalores: Os autovalores da matriz de Hilbert se distribuem de forma exponencial, com um grande autovalor dominante e vrios autovalores muito pequenos. Essa distribuio exponencial contribui para a sensibilidade da matriz a perturbaes.
- Acumulao de erros: Ao resolver um sistema linear com uma matriz mal condicionada, os erros de arredondamento se acumulam rapidamente durante as operaes computacionais. Isso significa que mesmo pequenas perturbaes nos dados podem levar a erros significativos na soluo.

#### Consequncias do mal condicionamento

- Dificuldade na resoluo numrica: Mtodos numricos padro, como o mtodo de Gauss-Seidel e Jacobi, podem apresentar resultados imprecisos ou at mesmo falhar completamente ao serem aplicados a matrizes de Hilbert mal condicionadas.
- Perda de preciso: A preciso dos resultados numricos pode ser significativamente reduzida, mesmo com o uso de mtodos sofisticados e tcnicas de mitigao de erros como o Sobre-relaxamento.

### Consideraes finais

A escolha do mtodo ideal para um sistema linear especfico depende de diversos fatores, como a preciso desejada, a convergncia esperada, o desempenho computacional disponvel e a experincia do pesquisador sobre o problema a ser modelado.  importante analisar as caractersticas do sistema e os requisitos da aplicao para tomar a melhor deciso.

```python
import numpy as np

def decomposicao_LU(matriz):
    n = len(matriz)
    # Initialize matrices L and U with zeros
    L = np.zeros((n, n))
    U = np.zeros((n, n))
    
    for i in range(n):
        # Matriz Triangular Superior
        for k in range(i, n):
            soma = 0
            for j in range(i):
                soma += (L[i][j] * U[j][k])
            U[i][k] = matriz[i][k] - soma
        
        # Matriz Triangular Inferior
        for k in range(i, n):
            if i == k:
                L[i][i] = 1  # Elementos da Diagonal de L que so 1
            else:
                soma = 0
                for j in range(i):
                    soma += (L[k][j] * U[j][i])
                L[k][i] = (matriz[k][i] - soma) / U[i][i]
    
    return L, U

def calcular_inversa(L, U):
    n = len(L)
    # Iniciando a matriz inversa com zeros
    matriz_inversa = np.zeros((n, n))

    # Verificar se a matriz  invertvel
    # Determinante muito prximo de zero
    if np.round(np.linalg.det(U), decimals=10) == 0:
        raise ValueError("A matriz no  invertvel.")
        
    
    # Resolvendo LY = I para Y (substituicao progressiva)
    for i in range(n):
        Y = np.zeros(n)
        Y[i] = 1
        for j in range(i):
            Y[i] -= L[i][j] * Y[j]
        Y[i] /= L[i][i]
        
        # Resolvendo UX = Y para X (substituio regressiva)
        for j in range(n-1, -1, -1):
            matriz_inversa[j][i] = Y[j]
            for k in range(j+1, n):
                matriz_inversa[j][i] -= U[j][k] * matriz_inversa[k][i]
            matriz_inversa[j][i] /= U[j][j]
    return matriz_inversa

A1 = np.array([[10, 2, -1],
              [-3, -6, 2],
              [1, 1, 5]])

print("\nMatriz A1:")
print_matriz(A1)
L, U = decomposicao_LU(A1)
print("\nMatriz L:")
print_matriz(L)
print("\nMatriz U:")
print_matriz(U)

matriz_inversa = calcular_inversa(L, U) # Invertivel
print("\nMatriz Inversa:")
print_matriz(matriz_inversa)

print("\nResultado de A * A")
print(np.round(np.dot(A1, matriz_inversa), decimals=1))
print("--------------------------")
```
